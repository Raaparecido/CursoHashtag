### Como foi a evolução das exportações para a frança ao longo dos anos?

Crie uma tabela com os anos de 2016, 2017, 2018, 2019 e 2020 e o valor US$ de exportações

```python
#Filtrando antes do groupby
tabela_anos_exportacao = tabela[['Year','US$ FOB']].groupby('Year').sum()
#display(tabela_anos_exportacao)

#Filtrando depois
"""tabela_anos_exportacao = tabela.groupby("Year").sum(numeric_only=True)
tabela_anos_exportacao = tabela_anos_exportacao[["US$ FOB"]]
"""

#Função para formatar 
def formatar(valor):
    valor_formatado = f'US$ {valor:,}'
    return valor_formatado

#Aplicando a Função no condigo
tabela_anos_exportacao['US$ FOB'] = tabela_anos_exportacao['US$ FOB'].apply(formatar)

display(tabela_anos_exportacao)
```

### Quais os produtos mais exportados ao longo de todo o período?

Queremos saber os produtos com maior valor US$ de exportação

```python
valor_total_expotacao = tabela[['SH2 Description', 'US$ FOB']].groupby('SH2 Description').sum()
valor_total_expotacao = valor_total_expotacao.sort_values(by='US$ FOB', ascending=False)

valor_total_expotacao['US$ FOB'] = valor_total_expotacao['US$ FOB'].apply(formatar)

display(valor_total_expotacao)

```

### Em 2020 qual cidade mais exportou para a França?
Em valores US$

```python
tabela2020 = tabela.loc[tabela['Year']==2020, :]

tabela_cidade_2020 = tabela2020[['City', 'US$ FOB']].groupby('City').sum()
tabela_cidade_2020 = tabela_cidade_2020.sort_values(by='US$ FOB', ascending=False)

tabela_cidade_2020['US$ FOB'] = tabela_cidade_2020['US$ FOB'].apply(formatar)

display(tabela_cidade_2020)
```

### Quais os produtos mais exportados (em US$) que as 2 maiores cidades (em exportação em 2020) exportaram?

```python
cidade = tabela_cidade_2020.index[3]
tabela_cidade = tabela2020.loc[tabela2020['City']==cidade, :]

tabela_cidade = tabela_cidade[['SH2 Description', 'US$ FOB']].groupby('SH2 Description').sum()
tabela_cidade = tabela_cidade.sort_values(by='US$ FOB', ascending=False)

tabela_cidade['US$ FOB'] = tabela_cidade['US$ FOB'].apply(formatar)

display(tabela_cidade)

```
